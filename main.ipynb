```python
results = model.predict(source="/kaggle/input/datasets/fariharayhanmim/motorcycle-violation/test/images")
```

```python
from collections import Counter

violation_counter = Counter()

total_images = len(results)

for r in results:
    boxes = r.boxes
    if boxes is not None:
        for box in boxes:
            class_id = int(box.cls)
            class_name = model.names[class_id]
            violation_counter[class_name] += 1

print("Total Images:", total_images)
print("Violation Counts:", violation_counter)
```

```python
total_violations = sum(violation_counter.values())

for k, v in violation_counter.items():
    print(f"{k}: {round((v/total_violations)*100,2)}%")
```

```python
labels = violation_counter.keys()
values = violation_counter.values()

plt.figure()
plt.bar(labels, values)
plt.xticks(rotation=45)
plt.title("Motorcycle Violation Distribution")
plt.show()
```
